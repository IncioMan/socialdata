- **PdDistrict** has been encoded into a one-hot encoding. This is due to its nature not having an ordinal relationship between values. ### Fitting functions
We now fit a Random Forests Classifier in order to classify which crime is being committed. To this extend cross-validation is ensure to prevent overfitting and identification of optimal hyperparameters for the Random Forest method. The cross-validation strategy is a one-layer 3-fold cross-validation for validation of models, with 10% of the data saved for testing models and estimating generelization error.

For hyperparameter-tuning the n_estimaters, max_depth, min_samples_split and min_samples_leaf are considered. The n_estimaters indicate the number of trees in the forest, the max_depth is maximum depth for each tree, the min_samples_split are the minimum samples in a node required to perform a split and the min_samples_leaf indicate the minimum number of observations required for a node to be considered a leaf.

```python
random_grid = {'n_estimators': [50,100,500],
               'max_depth': [10,50,100],
               'min_samples_split': [2, 5, 10],
               'min_samples_leaf': [1, 2, 4],}
```

```python
def cv_classifier(X, y, weight=None):
    # Use the random grid to search for best hyperparameters
    # First create the base model to tune
    rf = RandomForestClassifier(class_weight=weight)
    # Random search of parameters, using 3 fold cross validation,  
    # search across 10 different combinations, and use all available cores 
    rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=2, random_state=42, n_jobs = -1)
    models = rf_random.fit(X, y)
    best_model = models.best_estimator_
    return best_model, models
```

```python
def predict_plot_confusion_matrix(clf, X_test, y_test, crime1, crime2, title, ax=None):
    y_predict = clf.predict(X_test)
    print_confusion_matrix(title, y_test, y_predict, crime1, crime2, ax=ax)
    return y_test, y_predict
```

```python
def print_confusion_matrix(title, y_test, y_predict, crime1, crime2, ax=None):
    cf_matrix = confusion_matrix(y_test, y_predict)
    group_names = [f"True {crime1}",f"False {crime1}",f"False {crime2}",f"True {crime2}"]
    group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    if not ax:
        ax =  sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
    else:
        ax =  sns.heatmap(cf_matrix, annot=labels, ax=ax, fmt='', cmap='Blues')
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    ax.set_title(title)
```

We take the data for the period of 2010-2018 and group it by hour-of-the-day.

```python
selected_data = crimedata[(crimedata["DateTime"].dt.year > 2009) & (crimedata["DateTime"].dt.year < 2019)]
selected_data['hour_of_the_day'] = selected_data.DateTime.dt.hour
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

In order to easily compare how the distribution of crimes differ from each other, not absolute numbers, we normalize the data within each crime category and rearrange the category.

```python
hourlycrime = selected_data.groupby(['hour_of_the_day', 'Category']).size() / selected_data.groupby(['Category']).size() # Normalization
hourlycrimedf = hourlycrime.unstack()
hourlycrimedf['hour'] = hourlycrimedf.index
hourlycrimedf = hourlycrimedf.reset_index()
```

```python
# Enables the figure to be an output in the notebook
output_notebook()
```

In order to make an interactive bar chart for the crime data in the period of 2010 to 2019, the bokeh package is utilized. It should be noted that the data is normalized in relation to the total number of crimes for each category, therefore the figure will illustrate the relative occurence of crimes for each cateogry, rather than the total number of crimes. 

We follow five steps:

1. Cast the data to a Bokeh ColumnDataSource object as this format is what bokeh utilizes.

```python
source = ColumnDataSource(hourlycrimedf)
```

2. Create the figure object to place the various bar charts for each crime cateogry in.

```python
hours = list(map(str, np.array(range(24))))
p = figure(x_range=hours, plot_width = 900, plot_height = 450, title="Hourly relative crimerates in San Fransisco in 2010-2019",
           toolbar_location=None, tools="", x_axis_label = 'Hour of the day', y_axis_label = 'Ratio')
```

3. Iterate through the different crime-types and add each as a bar chart to the figure object.

The bar_colors attribute stores the colors for each of the different focuscrimes.
The top variable indicates the height of the bars and the dodge variable function ensures that the bars are centered around the correct time. The alpha value indicates the transparency for the colors in the bar chart.

```python
bar_colors=viridis(len(focuscrimes))
bar = {} # to store vbars
for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x=dodge('hour', 0.5, range=p.x_range),  top = i, source = source, 
                    color = bar_colors[indx], width = 0.6, alpha = 0.9) 
```

4. Add the legend to the figure.

```python
items =[]
for i in focuscrimes:
    items.append((i, [bar[i]]))
legend = Legend(items = items, location = (0,-30))
p.add_layout(legend, 'right') 
```

5. Make the legend interactive to the user. Clicking categories in the legend results in showing/hiding whichever category is selected or de-selected.

```python
p.legend.click_policy = "hide"
show(p)
```
