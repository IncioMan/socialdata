* **What's the point of Figure 7?**

    Figure 7 investigates the *visual narrative* and *narrative structure* choices of 58 visualizations belonging to different genres. Firstly, the authors found that different narrative formats use different ordering structures. For instance, slide shows tend to use a linear ordering prescribed by the authors, while annotated graphs tend to define a user-directed path.
    Secondly, while some interactivity techniques (e.g. details-on-demand) are extensively used and appear in the great majority of the visualizations, some other techniques are very rarely used. For instance, *tacit tutorials*, that seamlessly enable the user to familiarize with the interactive features, are only used by 6 visualizations.
    Thirdly, while headlines and annotations are frequently used, other messagging techniques are under-utilized. For instance, *multi-messaging techniques* (e.g. text and images working together), which strengthen the narrative, are mainly used in slideshows and videos. *Summaries* and *repetitions* are also only heavily under-utilized, although repetition is key for an effective communication to have long-lasting effect.
     


* **Use Figure 7 to find the most common design choice within each category for the Visual narrative and Narrative structure (the categories within visual narrative are 'visual structuring', 'highlighting', etc).**

    The *visual narrative* techniques are divided into three categories:
    * Visual structuring: The most common technique is consistent visual platform, followed by establishing shot/splash screen and progress bar/timebar
    * Highlighting: The most common techniquer is feature distinction, followed by close-up and motions
    * Transition guidance: The most common techniques is object continuity, followed by animated transitions
   
  The *narrative structure* tactics are also divided into three categories:
    * Ordering: The most common technique is user directed path, followed by linear ordering
    * Interactivity: The most common technique is filtering/selection/search, followed by hover highlighting/details
    * Messagging: The most common technique is captions/headlines, followed by annotations

* **What are the three key elements to keep in mind when you design an explanatory visualization?**

    The three elements to keep in mind are:
    * Start with a question: what is that you want to communicate, what do you want people to take away?
    * Allow exploration: let the users investigate. By interacting, the audience will be more engaged and, thus, will learn more. The exploration can be encouraged by giving an overview of the data, by allowing viewers to zoom and filter on the data and by allowing details on demand.
    * Know your readers: design for an audience, considering its background knowledge, its knowledge of the dataset, its expected engagement and effort.

* **Explain in your own words: How is explanatory data analysis different from exploratory data analysis?**

*Exploratory data analysis* aims at analysing a data set to familiarize with it, summarize its characteristics, formulate hypotheses, support the selection of appropriate statistical techniques, or further collect data.
*Explanatory data analysis* focuses on the communication of the relevant aspects of the data set, by presenting interpretable patterns or models.
