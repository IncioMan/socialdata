5. *What is a Quantile plot? What is it good for.*

A quantile plot is the CDF plot in which the x and y axes have been switched. A quantile plot is useful if we have a specific percentage X in mind and we want to quickly know which value is not exceeded by that percentage X of the data set.

6. *How is a Probablity plot defined? What is it useful for? Have you ever seen one before?*

A Probability plot is a plot that provides an intutive way to determine wether some data are distributed according to a specific theoretical distribution (such as the Gaussian distribution). If the data points fall onto a straight line resulting from the inverse transformation of the theoretical distribution, then the data points behave like the theoretical distribution.

7. *One of the reasons I like DAOST is that Janert is so suspicious of mean, median, and related summary statistics. Explain why one has to be careful when using those - and why visualization of the full data is always better.*

Generally, summary statistics are values that summarize the average location of the data points and their dispersion. This summarization entails a big information loss. Moreover, these statistics should only be used when dealing with unimodal distriibutions (with a single, central peak). If the distribution is skewed or has outliers, the mean and the standard deviation should be avoided.

8. *I love box plots. When are box plots most useful?*

Since box plots are good to quickly visualize how the data points are distributed (simmetry, cntral peak and tails), they are most useful when comparing several distribution among one another.

9. *The book doesn't mention violin plots. Are those better or worse than box plots? Why?*

Violine plots include both the summary statistics presented by a boxplot and the distribution displayed by a KDE plot. For this reason, violin plots, can be particularly useful when dealing with multimodal data (a distribution with more than one peak). Moreover, when the data set is big and there are a lot of outliers, the outliers displayed in a box plot would overlap and result in an ineffective visualization. This problem does not occure with violin plots.

6. *OK, we're almost done, but we need some box plots. Here, I'd like you to use the box plots to visualize fluctuations of how many crimes happen per day. We'll use data from the 15 focus crimes defined last week.For the full time-span of the data, calulate the number of crimes per day within each category for the entire duration of the data.Create a box-and whiskers plot showing the mean, median, quantiles, etc for all 15 crime-types side-by-side. There are many ways to do this. I like to use matplotlibs's built in functionality. What does this plot reveal that you can't see in the plots from last time?*

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
df4 = crimedata[crimedata['Category'].isin(focuscrimes)]
df4 = df4.groupby(["Category", "Date"]).count().IncidntNum

fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
ax.boxplot([df4[i] for i in focuscrimes], vert=False)
ax.set_yticklabels(focuscrimes)
ax.set_title("Crimes per day")
```

The plot reveals that, in addition to occurring with different frequencies, some crime categories fluctuate much more than others from day to day. For instance, drug/narcotic and vehcile theft have similar median as burglary and vandalism, but a larger interquartile range and more extreme outliers. This is due to the fact that those kind of crimes are more likely to occur in some specific days of the year.

7. Also I want to show you guys another interesting use of box plots. To get started, let's calculate another average for each focus-crime, namely what time of day the crime happens. So this time, the distribution we want to plot is the average time-of-day that a crime takes place. There are many ways to do this, but let me describe one way to do it.
For datapoint, the only thing you care about is the time-of-day, so discard everything else.
You also have to deal with the fact that time is annoyingly not divided into nice units that go to 100 like many other numbers. I can think of two ways to deal with this.
For each time-of-day, simply encode it as seconds since midnight.
Or keep each whole hour, and convert the minute/second count to a percentage of an hour. So 10:15  10.25, 8:40  8.67, etc.
Now you can create box-plots to create an overview of when various crimes occur. Note that these plot have quite a different interpretation than ones we created in the previous exercise. Cool, right?

```python
df5 = crimedata[crimedata['Category'].isin(focuscrimes)]
df5['Time'] = pd.to_datetime(df5['Time'])#.dt.minute
df5['SecondsSinceMidnight']= df5['Time'].dt.hour * 60*60 + df5['Time'].dt.minute *60
```

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
ax.boxplot([df5[df5['Category']== i].SecondsSinceMidnight for i in focuscrimes], vert=False)
ax.set_yticklabels(focuscrimes)
ax.set_title("Distribution of crimes over the day")
```
